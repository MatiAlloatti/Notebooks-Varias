```python
from fastai.vision import *
from fastai import *
import glob, os

root_dir = "/storage/notebooks/"
base_dir = root_dir + 'Bears/'
path = Path(base_dir)
classes = ['teddy','grizzly','black','polar','panda']
print(path)
```

```python
folder1 = 'black'
file = 'bBears.csv'
dest = path/folder1
dest.mkdir(parents=True, exist_ok=True)
download_images(path/file, dest, max_pics=200)
folder2 = 'teddy'
file = 'tBears.csv'
dest = path/folder2
dest.mkdir(parents=True, exist_ok=True)
download_images(path/file, dest, max_pics=200)
folder3 = 'grizzly'
file = 'gBears.csv'
dest = path/folder3
dest.mkdir(parents=True, exist_ok=True)
download_images(path/file, dest, max_pics=200)
folder4 = 'polar'
file = 'plrBears.csv'
dest = path/folder4
dest.mkdir(parents=True, exist_ok=True)
download_images(path/file, dest, max_pics=200)
folder5 = 'panda'
file = 'pdBears.csv'
dest = path/folder5
dest.mkdir(parents=True, exist_ok=True)
download_images(path/file, dest, max_pics=200)
```

```python
for c in classes:
    print(c)
    verify_images(path/c, delete=True, max_size=500)
```

```python
np.random.seed(42)
data = ImageDataBunch.from_folder(path, train=".", valid_pct=0.2,
        ds_tfms=get_transforms(), size=224, num_workers=4).normalize(imagenet_stats)
```

```python
data.classes
#data2.classes
```

```python
data.show_batch(rows=5, figsize=(14,14))
#data2.show_batch(rows=5, figsize=(14,14))
```

```python
data.classes, data.c, len(data.train_ds), len(data.valid_ds)
#data2.classes, data2.c, len(data2.train_ds), len(data2.valid_ds)
```

```python
learn = cnn_learner(data, models.resnet34, metrics=error_rate)
# learn = cnn_learner(data2, models.resnet34, metrics=error_rate)
```

```python
learn.fit_one_cycle(5)
```

```python
learn.save('stage-1', return_path=True)
```

```python
learn.load('stage-1');
```

```python
learn.unfreeze()
```

```python
learn.lr_find()
```

```python
learn.recorder.plot()
```

```python
learn.fit_one_cycle(2, max_lr=slice(1e-5,1e-4));
# learn.fit_one_cycle(2, max_lr=slice(6e-5,3e-4))
```

```python
learn.save('stage-2')
#learn.save('stage-2b')
```

```python
learn.load('stage-2');
```

```python
interp = ClassificationInterpretation.from_learner(learn)
```

```python
interp.plot_confusion_matrix()
```
